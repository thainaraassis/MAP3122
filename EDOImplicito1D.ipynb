##EDO unidimensional por Euler Implícito

Resolveremos a equação y' = 2y utilizando Euler Implícito.

```python
def solution(n,a,b,y0):

  y = np.empty((n+1))
  y[0] = y0

  step = (b-a)/n

  for i in range(1,n+1):
   # y[i] = y[i-1] + step*2*y[i-1] #Euler explícito
   y[i] = y[i - 1] / (1 - 2 * step) # Euler implícito

  return y
```

## Tabela de convergência
Uma vez definida a função solução, podemos resolver para diversos tamanhos de passos e obter uma tabela de convergência, pronta para ser usada em LaTeX.

```python
#Define o intervalo de análise e a condição inicial do problema
a =  0
b =  1
y0 = 1

ant = 0 ## erro anterior
n   = 4

for i in range (15):

    result = solution(n,a,b, y0)
    h = (b-a)/n
    erro = np.absolute(np.exp(2*b) - result[-1])

    if(i == 0):
        print(f'{n:6}', " & ", f'{h:e}', " & ", f'{erro:e}', " & ", "------------")

    else:
        ordem = np.absolute(np.log2(np.absolute(erro / ant)))
        print(f'{n:6}', " & ", f'{h:e}', " & ", f'{erro:e}', " & ", f'{ordem:e}')

    n *= 2
    ant = erro
```
