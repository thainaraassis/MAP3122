```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
import plotly.graph_objects as go
```

```python
def sen(x):
    return math.sin(x)

def atan(x):
    return math.atan(x)

def log(x):
    return math.log(x)

def getdata(n,init,end,mode):
    x = np.linspace(init,end,n)
    
    df = pd.DataFrame({'x' : x})

    if mode == 'sen':
        df['y'] = df['x'].apply(sen)
    elif mode == 'atan':
        df['y'] = df['x'].apply(atan)
    elif mode == 'log':
         df['y'] = df['x'].apply(log)
    else:
        return None
    return df
    


def plotgraph(dfs):
    for df in dfs:
        sns.lineplot(data = df, x = 'x', y = 'y')

def main():
    dfs = []
    init = 0
    end = 10
    mode = 'sen'
    for n in range (5,100,1):
        dfs.append(getdata(n,init,end,mode))
    plotgraph(dfs)
```

```python
main()
```

```python
def sen(x):
    return math.sin(x)

def atan(x):
    return math.atan(x)

def log(x):
    return math.log(x)

def getdata(n,init,end,mode):
    x = np.linspace(init,end,n)
    
    df = pd.DataFrame({'x' : x})

    if mode == 'sen':
        df['y'] = df['x'].apply(sen)
    elif mode == 'atan':
        df['y'] = df['x'].apply(atan)
    elif mode == 'log':
         df['y'] = df['x'].apply(log)
    else:
        return None
    
    return df
    


def plotgraph(dfs):
    linewidths = np.linspace(0,2,len(dfs))
    fig = go.Figure()
    for df, linewidth in zip(dfs, linewidths):
        fig.add_trace(go.Scatter(x=df['x'], y=df['y'], mode='lines', line=dict(width=linewidth)))
    fig.show()

def main():
    dfs = []
    init = 0
    end = 6.28
    mode = 'atan'
    for n in range (3,100):
        dfs.append(getdata(n,init,end,mode))
    plotgraph(dfs)
```
