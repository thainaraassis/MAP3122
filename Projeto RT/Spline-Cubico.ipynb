```python
import numpy as np
import matplotlib.pyplot as plt

def runge_kutta_4(f, x0, y0, xn, n):
    h = (xn - x0) / n
    x = np.linspace(x0, xn, n+1)
    y = np.zeros(n+1)
    y[0] = y0
    for i in range(1, n+1):
        k1 = h * f(x[i-1], y[i-1])
        k2 = h * f(x[i-1] + 0.5 * h, y[i-1] + 0.5 * k1)
        k3 = h * f(x[i-1] + 0.5 * h, y[i-1] + 0.5 * k2)
        k4 = h * f(x[i-1] + h, y[i-1] + k3)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return x, y

# Definindo a nova função da EDO, y' = 30*cos(x)*e^(yx)
def f(x, y):
    return 2 * np.cos(x) * np.exp(y * x)

# Condições iniciais e parâmetros
x0 = 0
y0 = 1 # Este valor é arbitrário e deve ser escolhido conforme o problema
xn = 10 # Este valor define até onde queremos aproximar a solução
n = 100 # Número de passos

# Executa o método de Runge-Kutta
x_rk, y_rk = runge_kutta_4(f, x0, y0, xn, n)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_rk, y_rk, 'b-', label='Runge-Kutta 4th Order Solution')
plt.title('Solution of y\' = 30cos(x)e^(yx) with Runge-Kutta 4th Order Method')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
