### Equação que se busca aproximar: y' = 2y com y(0) = 3

```python
import pandas as pd 
import numpy as np 
import math
import plotly.graph_objects as go
```

```python
index = [0]
real_value = [1]
aproximated_value = [1]
valor_init = 1


for i in range(1,100):
    index.append(i*0.01)
    real_value.append( (1*math.e) ** (2*i*0.01))

i = 0

def euler_recursivo(i,valor_ant):
    if i == 99:
        pass
    else:
        valor_atual = valor_ant + 0.01*2*valor_ant
        aproximated_value.append(valor_atual)
        euler_recursivo(i+1,valor_atual)

euler_recursivo(0,valor_init)
```

```python
df = pd.DataFrame({'index':index,'real_value':real_value,'aproximated_value':aproximated_value})
```

```python
df['error'] = df['real_value'] - df['aproximated_value']
```

```python
df.head(100)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=index, y=df['real_value'], mode='lines', name='real_value'))
fig.add_trace(go.Scatter(x=index, y=df['aproximated_value'], mode='lines', name='approximated_value'))
fig.add_trace(go.Scatter(x=index, y=df['error'], mode='lines', name='error'))

# Customize layout
fig.update_layout(title='Two Variables Line Plot',
                  xaxis_title='X Axis Label',
                  yaxis_title='Y Axis Label')

# Show plot
fig.show()
```
